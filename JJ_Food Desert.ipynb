```python
desert_csv_df.head()

outlier = desert_csv_df.groupby('State').mean()

outlier.reset_index(inplace = True)

outlier_df = outlier[['State', '% Change in Obesity', '% change in Access', '% change FF Per Capita', 'Median Household Income 2014']]

outlier_df.head
```

```python
#Graph % change obesity and Access
#Access:  if it is positive, means that access decrease, if negative access increases
#Obesity:  if positive, means they are fatter
#Expectation:  negative values in access result in negative values in obesity

import seaborn as sns
sns.set()

# Household income against Limited Access
w = sns.lmplot(x="% change in Access", y="% Change in Obesity", truncate=True, size=7, data=outlier_df)
plt.show()

#save png
plt.savefig("Access_v_Obesity_chnge.png")
```

```python

outlier_df['outlier'] = outlier_df.apply((lambda x: (x[2]> 0 and x[1]>0) or (x[2]< 0 and x[1]<0)  ), axis = 1) 
outlier_df.head()


outlier_df.to_csv("Outliers.csv")
```

```python
import seaborn as sns
sns.set()

# Household income against Limited Access
w = sns.lmplot(x="% change in Access", y="% Change in Obesity", truncate=True, size=7, data=outlier_df)
plt.show()

#save png
#plt.savefig("Access_v_Obesity_chnge.png")

```

```python
#False is Outlier

import seaborn as sns
sns.set()

size = np.arange(0, 1000, 10)

# Plot tip as a function of toal bill across days
g = sns.lmplot(x="% Change in Obesity", y="Median Household Income 2014", hue="outlier", 
               scatter_kws={"s": size, 'alpha':0.50,}, data=outlier_df, fit_reg=False)
               

#plt.ylim(15, 45)
#plt.xlim(0, 40)
plt.xlabel('Change in Obesity')
plt.ylabel('Change in FF')

plt.title('Outlier Compared to Standard')

plt.show()
```

```python
#obesity vs incomce 2009
fig, ax1 = plt.subplots()
x1 = outlier_df['% Change in Obesity']
y1 = outlier_df['% change in Access']
x2 = outlier_df['% Change in Obesity']
y2 = outlier_df['% change in Access']
x3 = outlier_df['% Change in Obesity']
y3 = outlier_df['% change in Access']

ax1.scatter(x1,y1, c= "blue", alpha=0.4, label = "MT")
ax1.scatter(x2,y2, c= "lime", alpha=0.4, label = "GA")
ax1.scatter(x3,y3, c= "red", alpha=0.4, label = "MA")
ax1.grid(True)

plt.title("Adult Obesity % vs. Median Household Income 2009")
plt.xlabel("Adult Obesity %")
plt.ylabel("Median Household Income")
plt.legend()
plt.show()
```
